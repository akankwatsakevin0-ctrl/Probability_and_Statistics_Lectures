```python
import numpy as np

# Jupyter cell (new file)
import matplotlib.pyplot as plt

x = np.array([1, 3, 5, 8, 9, 11, 13, 15])
y = np.array([36, 32, 27, 23, 21, 15, 14, 11])

m, b = np.polyfit(x, y, 1)           # slope and intercept
y_pred = m * x + b
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r2 = 1 - ss_res / ss_tot

xx = np.linspace(x.min() - 1, x.max() + 1, 200)
yy = m * xx + b

plt.figure(figsize=(8,5))
plt.scatter(x, y, color='C0', label='data points')
plt.plot(xx, yy, color='C1', label=f'best-fit: y = {m:.3f}x + {b:.3f}\n$R^2$ = {r2:.3f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot with line of best fit')
plt.legend()
plt.grid(True)
plt.show()
```
