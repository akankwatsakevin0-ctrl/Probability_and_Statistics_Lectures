### Correlation Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data ={
    "X":pd.Series([1,3,5,8,9,11,13,15,16,17,19]),
    "Y":pd.Series([36, 32, 27, 23, 21, 15, 14, 11, 12, 8, 5]),
}

df= pd.DataFrame(data)
df
```

```python
sns.scatterplot(x="X", y ="Y", data=df )
plt.show()
```

- Adding the line of Best fit

```python
sns.regplot(x="X", y ="Y", data=df )
plt.show()
```

```python
housing = pd.read_csv("Housing.csv") ## in the dir
housing.head(10)
```

## extract only numerical column

```python
numerical_df = housing.select_dtypes(include=['int64', 'float64'])

# numerical_df = housing.select_dtypes(exclude="object")
```

```python
numerical_df
```

- correlation matrix

```python
plt.figure(figsize=(6,6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.scatterplot(x="price", y ="parking", data=numerical_df )
plt.show()
```

## Pearson

```python
data ={
    "Math":pd.Series([84, 72, 67, 81]),
    "Hist":pd.Series([46, 48, 53, 19]),
}

df= pd.DataFrame(data)
df
```

```python
df.corr(method="pearson") 
```

## Spearman

```python
data ={
    "Math":pd.Series([84, 72, 67, 81]),
    "Hist":pd.Series([46, 48, 53, 19]),
}

df= pd.DataFrame(data)
df
```

```python
df.corr(method="spearman") 
```
