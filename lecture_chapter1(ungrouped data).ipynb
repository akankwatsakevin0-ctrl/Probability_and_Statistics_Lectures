#### Example dataset for ungrouped data

```python
## ages of students in BsIT 1:2
data = [23, 19, 20, 18, 22, 22, 23, 24, 28]
```

### Measures of Central Tendency.
- Mean (Average)
- Median (Middle value)
- Mode (Most frequent)

## Mean 
= summation of all x/ total number of x

```python
mean_value = sum(data)/len(data)
mean_value
```

```python
import numpy as np
np.mean(data)
```

```python
import pandas as pd
ds = pd.Series(data)

ds.mean()
```

## Median

```python
np.median(data)# numpy
```

```python
ds.median() ## pandas
```

## Mode

```python
from statistics import mode

mode(data)
```

```python
from statistics import multimode
multimode(data)
```

```python
ds.mode()
```

### data representation
>visually represent our data values

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(data, bins=6)
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.title("Histogram for ungrouped data")
plt.show()
```
